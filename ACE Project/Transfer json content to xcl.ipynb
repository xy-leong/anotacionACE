```python
import os
import json
import pandas as pd
```

```python
# Recover the path
directory_path = os.getcwd()
# root_path = os.path.dirname(directory_path)
print(directory_path)
# print(root_path)
```

```python
ace_terms_filepath = os.path.join(directory_path, "ace_terms_dictionary.json")
print(ace_terms_filepath)
```

```python


with open(ace_terms_filepath, "r", encoding="utf-8") as f:
    data = json.load(f)

rows = []
for key, terms in data.items():
    for term in terms:
        rows.append({"key": key, "term": term})

df = pd.DataFrame(rows)
df.to_excel("terms.xlsx", index=False)
```
