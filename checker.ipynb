```python
import os
import re
from pathlib import Path
```

```python
def parse_ann_file(ann_path):
    entities = []

    with open(ann_path, "r", encoding="utf-8") as f:
        for line in f:
            if not line.startswith("T4"):
                continue

            parts = line.strip().split("\t")
            print(parts)
            if len(parts) < 3:
                continue

            tag_info = parts[1].split()
            print(tag_info)
            label = tag_info[0]

            # --- ONLY ACE ENTITIES
            if not label.startswith("ACE_"):
                continue

            start = int(tag_info[1])
            end = int(tag_info[2]) if len(tag_info[1:]) == 2 else int(tag_info[-1])

            entities.append({
                "start": start,
                "end": end,
                "label": label
            })
            print(entities)

f = os.path.join("C:/Users/ASUS/Berlin/Practicas/ACE/brat/data/ACE_v1", "caso_2173316.ann")
parse_ann_file(f)
```
